```python
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
table = pd.read_csv('../data/VAF_evolution_whole_blood.csv').fillna(0)
table['timepoint'] = ['1','2','3']
table.index = table['timepoint'] + '\n' + table['MPN Clinical stage ']

table.drop(columns=['timepoint', 'MPN Clinical stage '], inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(4,3))
for mutation in table.columns:
    ax.plot(table.index, table[mutation], marker='o', label=mutation)

ax.set_ylabel('Whole blood bulk VAF (%)')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig('../plots/bulk_VAF_evolution.pdf', bbox_inches='tight')
```

```python
import pandas as pd

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt

import scipy.stats as st
import numpy as np
```

```python
likelihoods_gapfill_betaine = pd.read_csv('../output/gapfill_betaine_1plex_summary.csv', index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(2,4))
data = likelihoods_gapfill_betaine['mean_counts'].copy()


alpha = 0.3

bp = ax.boxplot(
    data,
    tick_labels=[
        'GIFT\n+ Betaine'
    ],
    patch_artist=True,
    boxprops=dict(facecolor='lightgrey', color='k',linewidth=1.5),
    medianprops=dict(color='k', linewidth=2),
    whiskerprops=dict(color='k', linewidth=1.5),
    capprops=dict(color='k', linewidth=1.5),
    showfliers=False, widths=0.7
)
# Color boxes: grey for no betaine, teal for betaine
bp['boxes'][0].set_facecolor('teal')

ax.set_ylabel('Mean counts per cell')
plt.tight_layout()


## print numbers to report
print('n=' + str(len(data)))
print("Median (Gapfill + Betaine):", np.median(data))

ax.set_ylim(-0.1,8)

plt.savefig('../plots/gapfill_betaine_count_boxplot.pdf')
```

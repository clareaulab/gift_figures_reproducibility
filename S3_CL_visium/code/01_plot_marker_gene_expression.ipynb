```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import matplotlib as mpl

matplotlib.rcParams['pdf.fonttype'] = 42

import sys
sys.path.append('../../1_figure_CL_proof_of_concept/code/')
from spatial_utils import read_gapfill_data, plot_marker_gene_spatial

import utils_00 as gf_utils
large_data_dir = gf_utils.large_data_dir
```

```python
data_path = f"{large_data_dir}gf_CL_VisiumHD/"
# Load data:
gapfill_sdata = read_gapfill_data(gf_adata_path = f"{data_path}GapFill_CellLineMix_Visium_GF.h5ad", WTA_dir = f"{data_path}GapFill_WTA.zarr", cores = 4)
```

```python
marker2color = {
    'PRAME': '#F79520',
    'ZNF595': '#F79520',
    'CPED1': '#F79520',
    'B2M': '#1C75BC',
    'IFI27': '#1C75BC',
    'MLLT3': '#39B54A',
    'AIF1': '#39B54A',
    'EIF1AY': '#39B54A'
}

for marker in [
    'PRAME', 'MLLT3', 'B2M', 'ZNF595', 'AIF1', 'IFI27', "CPED1", "EIF1AY"
]:
    color = marker2color.get(marker, 'gray')

    fig, ax = plot_marker_gene_spatial(
        gapfill_sdata,
        marker,
        color=color,
        resolution=16,
        figsize=(6, 6),
        title=f'{marker} Expression',
    )

    plt.tight_layout()
    plt.savefig(f'../plots/visium_marker_gene_expression_{marker}.pdf', bbox_inches='tight')
    plt.show()
```
